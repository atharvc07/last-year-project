# Exploratory Data Analysis (EDA) - Kaggle Dataset
In this notebook, we explore the Indian road accident dataset to understand patterns in severity based on various conditions.

**Goals:**
- Load and inspect data
- Handle any missing values
- Visualize Feature Distributions
- Identify trends (Weather, Time of Day, Alcohol Involvement)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plotting style for better visualizations
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

### 1. Load the Dataset

```python
# Load the dataset
df = pd.read_csv('../data/accident_prediction_india.csv')

# Clean column names (remove leading/trailing spaces)
df.columns = df.columns.str.strip()

# Display the first few rows
df.head()
```

### 2. Basic Dataset Information
Let's check the shape and missing values.

```python
print(f"Dataset Shape (Rows, Columns): {df.shape}")
print("\n--- Data Types & Non-Null Counts ---")
df.info()

print("\n--- Missing Values Check ---")
print(df.isnull().sum())
```

### 3. Target Variable Analysis: Accident Severity

```python
sns.countplot(x='Accident Severity', data=df, palette='viridis')
plt.title('Distribution of Accident Severity')
plt.xlabel('Severity Level')
plt.ylabel('Count')
plt.show()
```

### 4. Impact of Weather and Road Conditions

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Weather Impact
sns.countplot(ax=axes[0], y='Weather Conditions', hue='Accident Severity', data=df, palette='magma')
axes[0].set_title('Accident Severity by Weather Conditions')
axes[0].set_ylabel('Weather Conditions')

# Road Condition Impact
sns.countplot(ax=axes[1], y='Road Condition', hue='Accident Severity', data=df, palette='Set2')
axes[1].set_title('Accident Severity by Road Condition')
axes[1].set_ylabel('Road Condition')

plt.tight_layout()
plt.show()
```

### 5. Impact of Alcohol Involvement

```python
if 'Alcohol Involvement' in df.columns:
    sns.countplot(x='Alcohol Involvement', hue='Accident Severity', data=df, palette='coolwarm')
    plt.title('Accident Severity by Alcohol Involvement')
    plt.xlabel('Alcohol Involvement')
    plt.ylabel('Count')
    plt.show()
```

### 6. Time of Day Trends

```python
if 'Time of Day' in df.columns:
    sns.countplot(x='Time of Day', data=df, palette='plasma', order=df['Time of Day'].value_counts().index)
    plt.title('Accidents by Time of Day')
    plt.xlabel('Time of Day')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```
